```python
df_numeric = df.select_dtypes(include = ['float64','int64'])
df_numeric
```

```python
df_numeric = df_numeric.drop(['view','condition','waterfront','yr_built','yr_renovated'],axis = 1)
df_numeric
```

```python
#divide the data into the input and output
x = df_numeric.iloc[:,0:7].values
x
```

```python
y = df_numeric.iloc[:,0].values
y
```

```python
#train and test the varuables - train_test_split()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape) #(205,) - 100%
print(y_train.shape)#(153,) -75%
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=10)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test) #using the input values , we predict the output
y_pred
```
